# Example usage for dataset dSpritesBackgroundTime

### so far, only shapetype='circle' works

```python
import torch
from torch.utils.data import DataLoader
from torchvision import transforms

import matplotlib.pyplot as plt
import numpy as np
from dspritesbT import dSpriteBackgroundDatasetTime, show_movies_grid
```

### single image, n_timesteps = 5

```python
transformed_dataset = dSpriteBackgroundDatasetTime(transform=transforms.Resize((32,32)),
                                                  n_timesteps=5,
                                                  shapetype='circle')
```

```python
idx = 3001
sample_trans,latent_trans = transformed_dataset[idx]
# show_movies_grid expects batches in the 0-th dimension
show_movies_grid(sample_trans.unsqueeze(0))
```

### single image, n_timesteps = 5, min_travelled_distance = 1.0 (too high, only for visualization)

```python
transformed_dataset = dSpriteBackgroundDatasetTime(transform=transforms.Resize((32,32)),
                                                   min_travelled_distance = 1.0,
                                                   n_timesteps=5,
                                                   shapetype='circle')
```

```python
idx = 3001
sample_trans,latent_trans = transformed_dataset[idx]
# show_movies_grid expects batches in the 0-th dimension
show_movies_grid(sample_trans.unsqueeze(0))
```

### batch of images, n_timesteps = 10

```python
transformed_dataset = dSpriteBackgroundDatasetTime(transform=transforms.Resize((32,32)),
                                                  n_timesteps=10,
                                                  shapetype='circle')
```

```python
dataloader = DataLoader(transformed_dataset, batch_size=25,
                       shuffle=True, num_workers=1)
print('One minibatch, shuffled and scaled down')
for i,[samples,latents] in enumerate(dataloader):
   print('Minibatch {}, batch size: {}'.format(i,samples.size())) 
   show_movies_grid(samples)
   break
```
